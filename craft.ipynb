```python
import cv2
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
def find_binary_img_by_kmean(img):
    # Cover img to hsv
    img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    img_2d = img.reshape((-1,3))
    kmeans = KMeans(n_clusters=2).fit(img_2d)
    labels = kmeans.labels_
    cluster_img = labels.reshape(img.shape[:2])

    return cluster_img
```

```python
img = cv2.imread("samples/3.jpg")
cluster_img = find_binary_img_by_kmean(img)

plt.imshow(cluster_img, cmap="gray")
plt.show()
```

```python
import os
```

```python
for img_path in os.listdir('data\mit_dataset'):
    img = cv2.imread('data\mit_dataset/' + img_path)
    cluster_img = find_binary_img_by_kmean(img)
    plt.imshow(cluster_img, cmap="gray")
    plt.show()
    # write a gray image
    cv2.imwrite(f"data\mit_dataset_bin\{img_path}", cluster_img)
    break
```
